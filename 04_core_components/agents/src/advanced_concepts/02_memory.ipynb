### 💾 메모리 (Memory)

**메모리(Memory)**는 에이전트가 대화 기록 외에 추가적인 정보를 저장하고 추적할 수 있도록 하는 고급 개념입니다.</br>
이는 에이전트가 단기 및 장기적으로 문맥(Context)을 유지하며 더 복잡하고 연속적인 작업을 수행할 수 있도록 합니다.

#### 1. 기본 개념 및 대화 기록
에이전트는 LangGraph의 상태(State)를 통해 대화 기록을 자동으로 유지합니다.

+ 대화 기록: 상태 내의 메시지 시퀀스를 통해 자동으로 유지됩니다. 이는 에이전트가 단기 및 장기적으로 문맥(Context)을 유지하며 더 복잡하고 연속적인 작업을 수행할 수 있도록 합니다.에이전트가 이전에 무슨 말을 했고 어떤 도구를 사용했는지 등을 기억하는 기본적인 형태의 메모리입니다.

#### 2. 단기 메모리 (Short-term Memory)와 Custom State

문서에서 '메모리'는 주로 대화 기록을 넘어서는 추가 정보를 상태에 저장하는 단기 메모리로 사용됩니다.

+ 역할: 사용자 선호도, 세션 변수, 임시 계산 결과 등 대화 중 일시적으로 추적해야 하는 정보를 저장합니다.

+ 구현: 기본 AgentState를 확장하는 **사용자 정의 상태 스키마(Custom State Schema)**를 통해 구현됩니다. 이 스키마는 반드시 TypedDict 타입이어야 합니다.

#### 3. 장기 메모리 (Long-term Memory)

주로 세션 간에 지속되어야 하는 정보(예: 영구적인 사용자 프로필, 과거의 모든 상호 작용 기록)를 다룹니다.
