### 🤖 모델 (Models)

LangChain 프레임워크의 **핵심 추론 엔진**(reasoning engine)으로서 **LLM**(Large Language Model)을 의미합니다.</br>

단순히 텍스트를 생성하는 것 외에 다음과 같은 기능을 지원합니다:

+ Tool calling (도구 호출): 데이터베이스 쿼리나 API 호출 같은 외부 도구를 호출하고 그 결과를 답변에 활용.
+ Structured output (구조화된 출력): 모델의 응답이 정의된 형식(예: Pydantic 모델, JSON Schema)을 따르도록 제약.
+ Multimodality (다중 모드): 텍스트 외의 이미지, 오디오, 비디오와 같은 데이터를 처리하고 반환.
+ Reasoning (추론): 결론에 도달하기 위해 다단계 추론을 수행.

Models의 활용 방식은 두 가지입니다:

+ With agents (에이전트와 함께): 에이전트 생성 시 모델을 동적으로 지정하여 에이전트 루프 내에서 사용.
+ Standalone (독립적으로): 에이전트 프레임워크 없이 텍스트 생성, 분류, 추출 등과 같은 특정 작업을 위해 모델을 직접 호출.

#### ⚙️ 모델 매개변수 (Parameters)

| 매개변수 | 유형 | 필수 여부 | 설명 |
| :--- | :--- | :--- | :--- |
| **model** | 문자열 | 필수 | 제공업체에서 사용할 특정 모델의 **이름 또는 식별자**입니다. |
| **api_key** | 문자열 | 선택 | 모델 제공업체에 **인증**하는 데 필요한 키이며, 보통 환경 변수로 접근합니다. |
| **temperature** | 숫자 | 선택 | 모델 출력의 **무작위성**을 제어합니다. 높을수록 창의적이고, 낮을수록 결정적입니다. |
| **timeout** | 숫자 | 선택 | 요청을 취소하기 전에 모델 응답을 기다리는 **최대 시간** (초 단위)입니다. |
| **max_tokens** | 숫자 | 선택 | 응답에 포함될 수 있는 **총 토큰의 수**를 제한하여 출력 길이를 제어합니다. |
| **max_retries** | 숫자 | 선택 | 네트워크 문제 등으로 요청 실패 시 **재전송을 시도할 수 있는 최대 횟수**입니다. |

### 🚀 Invocation (호출)

모델이 출력을 생성하도록 요청하는 3가지 방법입니다.

| 호출 방법 | 설명 | 사용 목적 | 반환 형식 |
| :--- | :--- | :--- | :--- |
| **`invoke()`** | 단일 메시지 또는 메시지 목록을 입력으로 받아 **완전한 응답**을 생성합니다. | 간단한 텍스트 생성, 분류, 추출 등 **최종 응답 전체**가 필요한 경우. | 완성된 단일 메시지 (`AIMessage`) |
| **`stream()`** | 출력이 생성되는 **동시에 실시간으로 출력 조각**을 제공하는 이터레이터를 반환합니다. | 긴 응답에 대한 **사용자 경험 개선** 및 실시간 출력 처리가 필요한 경우. | 출력 조각 (`AIMessageChunk`)의 이터레이터 |
| **`batch()`** | 독립적인 요청들을 모아 모델에 **일괄적으로 전송**하여 병렬로 처리합니다. | 대량의 독립적인 요청을 **효율적으로 처리**하여 성능 향상 및 비용 절감이 필요한 경우. | 입력 순서에 따른 응답 목록 |

#### Invoke

```python
model = init_chat_model(
    "google_genai:gemini-2.5-flash-lite",
)
```

```python
response = model.invoke("앵무새는 왜 깃털이 화려할까요?")
print(response)
```

```python
conversation = [
    {"role": "system", "content": "당신은 영어를 한글로 번역하는 친절한 조수입니다."},
    {"role": "user", "content": "I love programming."},
    {"role": "assistant", "content": "프로그래밍을 사랑합니다."},
    {"role": "user", "content": "I love building applications."}
]

response = model.invoke(conversation)
print(response)
```

```python
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

conversation = [
    SystemMessage("당신은 영어를 한글로 번역하는 친절한 조수입니다."),
    HumanMessage("I love programming."),
    AIMessage("프로그래밍을 사랑합니다."),
    HumanMessage("I love building applications.")
]

response = model.invoke(conversation)
print(response)
```

#### Stream

```python
for chunk in model.stream("앵무새는 왜 깃털이 화려할까요?"):
    print(chunk.text, end="|", flush=True)
```

```python
full = None # 초기화 변수 추가
for chunk in model.stream("하늘은 무슨 색인가요?"):
    full = chunk if full is None else full + chunk
    print(full.text)
    print("-" * 100)

print(full.content_blocks) # 스트리밍이 최종적으로 invoke()와 동일한 결과물을 도출함을 확인할 수 있음
```

### Batch

```python
responses = model.batch([
    "앵무새는 왜 깃털이 화려한가요?",
    "비행기는 어떻게 하늘을 날까요?",
    "양자 컴퓨팅이란 무엇인가요?"
])
for response in responses:
    print(response)
```

### 🛠️ Tool Calling (도구 호출)

모델의 기능을 텍스트 생성을 넘어 외부 세계와의 상호작용으로 확장합니다.

```python
from langchain.tools import tool

@tool
def get_weather(location: str) -> str:
    """특정 위치의 날씨를 가져옵니다."""
    return f"It's sunny in {location}."


model_with_tools = model.bind_tools([get_weather]) # 사용자가 정의한 도구를 모델이 사용할 수 있도록 연결

response = model_with_tools.invoke("부산의 날씨는 어떤가요?")
for tool_call in response.tool_calls:
    # View tool calls made by the model
    print(f"Tool: {tool_call['name']}")
    print(f"Args: {tool_call['args']}")
```

### 🧱 Structured Outputs (구조화된 출력)

모델이 생성하는 응답이 사용자가 미리 정의한 특정 스키마(schema) 형식을 따르도록 제약하는 기능입니다.

#### Pydantic

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    """영화에 대한 세부 정보."""
    title: str = Field(..., description="영화의 제목")
    year: int = Field(..., description="영화가 개봉된 연도")
    director: str = Field(..., description="영화의 감독")
    rating: float = Field(..., description="10점 만점 영화 평점")

model_with_structure = model.with_structured_output(Movie)
response = model_with_structure.invoke("영화 기생충에 대한 세부 정보를 제공해줘")
print(response)
```

```python
from typing_extensions import TypedDict, Annotated

class MovieDict(TypedDict):
    """영화에 대한 세부 정보."""
    title: Annotated[str, ..., "영화의 제목"]
    year: Annotated[int, ..., "영화가 개봉된 연도"]
    director: Annotated[str, ..., "영화의 감독"]
    rating: Annotated[float, ..., "10점 만점 영화 평점"]

model_with_structure = model.with_structured_output(MovieDict)
response = model_with_structure.invoke("영화 기생충에 대한 세부 정보를 제공해줘")
print(response)
```

```python
import json

json_schema = {
    "title": "영화",
    "description": "영화에 대한 세부 정보.",
    "type": "object",
    "properties": {
        "title": {
            "type": "string",
            "description": "영화의 제목"
        },
        "year": {
            "type": "integer",
            "description": "영화가 개봉된 연도"
        },
        "director": {
            "type": "string",
            "description": "영화의 감독"
        },
        "rating": {
            "type": "number",
            "description": "10점 만점 영화 평점"
        }
    },
    "required": ["title", "year", "director", "rating"]
}

model_with_structure = model.with_structured_output(
    json_schema,
    method="json_schema",
)
response = model_with_structure.invoke("영화 기생충에 대한 세부 정보를 제공해줘")
print(response)
```
